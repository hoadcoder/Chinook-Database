```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

% matplotlib inline

import seaborn as sns
sns.set_style('darkgrid')
```

```python
# Read in A1.csv as a pandas dataframe

df1 = pd.read_csv('data/csv/A1.csv')
df1.head(10)
```

```python
# Define font size for titles

fontT = 13
```

```python
width = 0.09
alpha = 1

location = [0, 0.09, 0.18, 0.30, 0.42, 0.54, 0.66]
labels = ['2009', '2010', '2011', '2012', '2013']
colors = ['#CAF270', '#8DCD6D', '#5BA767', '#33805C', '#165B4B']

# Plot bars

plt.bar(location[0:3], df1.iloc[:3, 2], width, color=colors[0], alpha=alpha, label=labels[0]) # 2009 Bars
plt.bar(location[3], df1.iloc[3, 2], width, color=colors[1], alpha=alpha, label=labels[1]) # 2010 Bar
plt.bar(location[4], df1.iloc[4, 2], width, color=colors[2], alpha=alpha, label=labels[2]) # 2011 Bar
plt.bar(location[5], df1.iloc[5, 2], width, color=colors[3], alpha=alpha, label=labels[3]) # 2012 Bar
plt.bar(location[6], df1.iloc[6, 2], width, color=colors[4], alpha=alpha, label=labels[4]) # 2013 Bar

# Print the values within each bar

for a, b in zip(location, df1.iloc[:, 2]):
    plt.text(a, b - 1.5, '%.2f' % b, ha='center', va='top', color='white')

# Set up labels, title and legend   

plt.ylabel('Total gasto ($)', weight='bold', color='#112222')
plt.xlabel('ID do Cliente', weight='bold', color='#112222')
plt.xticks(location, df1.iloc[:, 1])
plt.title('Total gasto pelo cliente mais rent√°vel de cada ano', size=fontT)
plt.legend(loc='lower left', bbox_to_anchor=(1, 0))

# Save plot as png file

plt.savefig('imgs/A1.png', dpi=300, bbox_inches='tight');
```

```python
# Read in A2.csv as a pandas dataframe

df2 = pd.read_csv('data/csv/A2.csv')
df2.head()
```

```python
width = 1

location = np.arange(len(df2.iloc[:, 0]))
labels = df2.iloc[:, 0]
colors = ['#317256', '#398564', '#419873', '#49AB81', '#52BF90']

# Plot bars    
    
bars = plt.bar(location, df2.iloc[:, 1], width, color=colors)

# Print the values within each bar

for a, b in zip(location, df2.iloc[:, 1]):
    plt.text(a, b - 5, '%.2f' % b, ha='center', va='top', color='white')

# Set up labels, title and legend 

plt.ylabel('Capital gerado ($)', weight='bold', color='#112222')
plt.xlabel('Artistas', weight='bold', color='#112222')
plt.xticks(location, labels)
plt.title('Capital gerado por Artistas', size=fontT, color='#112222')

# Save plot as png file

plt.savefig('imgs/A2.png', dpi=300, bbox_inches='tight');
```

```python
# Read in A3.csv as a pandas dataframe

df3 = pd.read_csv('data/csv/A3.csv')
df3.head()
```

```python
people = df3.iloc[:, 0]
y_pos = 2 * np.arange(0, len(people), 1)
conversion_arr = df3.iloc[:, 1]

location = np.arange(len(df3.iloc[:]))
color = ['#52BF90']

# Define plot size

plt.figure(figsize=(8, 18), dpi=80)

# Plot horizontal bars

plt.barh(y_pos, conversion_arr, color=color, height=1)

# Print the values within each bar

for a, b in zip(conversion_arr, y_pos):
    plt.text(a + 3, b, '%.2f' % a, ha='left', va='center', color='black')

# Set up labels, title and legend 

plt.ylabel('Cliente', weight='bold', color='#112222')
plt.xlabel('Total gasto ($)', weight='bold', color='#112222')
plt.yticks(y_pos, people)
plt.title('Total gasto com os 5 artistas mais rent√°veis por Cliente', size=fontT, color='#112222')

# Invert y-axis

plt.gca().invert_yaxis()

# Save plot as png file

plt.savefig('imgs/A3.png', dpi=300, bbox_inches='tight');
```
